```python
import pandas as pd  
import numpy as np
# import re
import csv
from urllib.parse import urlparse
import urllib.request
```

```python
df = pd.read_csv("C:/Users/91635/Desktop/ml work/favorite-tweets usable.csv")
```

```python
df1= df.tail(51)
df1
```

```python
#finding URL from text/tweets
shorturl = []

pattern = r'(https?:\/\/(?:www\.)?[-a-zA-Z0-9@:%._+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}[-a-zA-Z0-9()@:%_+.~#?&/=]*)' 

df1["Text"]=df1["Text"].astype(str)

# df1['links'] = ''
#finding url and extracting patterns
df1['links']= df["Text"].str.extract(pattern, expand=True)
for s in df1['links']: 
    shorturl.append(s)


shorturl
```

```python
# Finding fill url
n = 0
u = 0
df1["links"]=df1["links"].apply(str)
Flist = []
for m in df1["links"]:
    if(m=="nan"):
        Flist.append("Nan")
        n = n + 1
    else:
        try:
            resp = urllib.request.urlopen(m)
            Flist.append(resp.url)
            u = u + 1
        except:
            Flist.append("URL ERROR")
```

```python
# full URL list
Flist
```

```python
#finding Domain
dlist =[]
for t in Flist:
    domain = urlparse(t).netloc
    dlist.append(domain)

dlist
```

```python

# Create a Pandas DataFrame.
#creating  a Dataframe 
URL_CSV = {
    'Short url':shorturl,
    'Full url' :Flist,
    'Domain list':dlist,
              }
URL = pd.DataFrame(URL_CSV)
```
